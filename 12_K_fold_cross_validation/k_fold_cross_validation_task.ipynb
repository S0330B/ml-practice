```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
from sklearn.datasets import load_iris
import numpy as np
iris=load_iris()
```

```python
from sklearn.model_selection import cross_val_score
```

```python
lr=cross_val_score(LogisticRegression(),iris.data,iris.target,cv=3)
lr
```

```python
np.average(lr)
```

```python
svm=cross_val_score(SVC(),iris.data, iris.target,cv=3)
```

```python
np.average(svm)
```

```python
rf=cross_val_score(RandomForestClassifier(),iris.data,iris.target,cv=3)
```

```python
np.average(rf)
```

```python
dt=cross_val_score(tree.DecisionTreeClassifier(),iris.data, iris.target, cv=3)
```

```python
np.average(dt)
```

```python
scores = {
    'Logistic Regression': np.average(lr),
    'Support Vector Machine': np.average(svm),
    'Random Forest': np.average(rf),
    'Decision Tree': np.average(dt)
}
sorted(scores.items(), key=lambda x: x[1], reverse=True)
```
